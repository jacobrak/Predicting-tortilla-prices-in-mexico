```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
```

```python
df = pd.DataFrame(pd.read_csv("tortilla_prices.csv"))
```

```python
# Changing dataframe to 01 dates of the month
df['Day'] = '01'
df['Date'] = pd.to_datetime(df[['Year', 'Month', 'Day']])
df = df[['Date', 'State', 'Price per kilogram']]
```

```python
df
```

```python
#Creating new df
unique_dates = df['Date'].unique()
new_df = []

for date in unique_dates:
    date_data = df[df['Date'] == date]
    
    mean_price = date_data['Price per kilogram'].mean()
    
    new_df.append({'Date': date, 'Mean_Price_per_kg': mean_price})
new_df = pd.DataFrame(new_df)
```

```python
new_df
```

```python
size = int(0.8*len(new_df))
train = new_df[new_df['Date'].dt.year < 2022]
test = new_df[new_df['Date'].dt.year >= 2022]
```

```python
train.head
```

```python
multiplicative_decomposition = seasonal_decompose(new_df['Mean_Price_per_kg'], model='multiplicative', period=30)
multiplicative_decomposition.plot()
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

additive_decomposition = seasonal_decompose(new_df['Mean_Price_per_kg'], model='additive', period=30)
additive_decomposition.plot()
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
train.head(100)
```

## Creating models

```python
warnings.filterwarnings("ignore")
time_series_data = train['Mean_Price_per_kg']

model = AutoReg(time_series_data, lags=20)
model_fit = model.fit()

print(model_fit.summary())

forecast = model_fit.forecast(steps=27)
plt.plot(test["Mean_Price_per_kg"])
plt.plot(forecast, label='Predicted')

plt.legend()
plt.show()
```

```python
model = SARIMAX(time_series_data, order=(1, 1, 5), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit()


forecast = model_fit.forecast(steps=27)
plt.plot(test["Mean_Price_per_kg"])
plt.plot(forecast, label='Predicted')
plt.legend()
plt.show()
```

```python
model = ARIMA(time_series_data, order =(1, 1, 4))
model_fit = model.fit()
print(model_fit.summary())

forecast = model_fit.get_forecast(steps=27)
y_hat = forecast.predicted_mean
yhat_conf_int = forecast.conf_int(alpha=0.05)

plt.plot(test["Mean_Price_per_kg"])
plt.plot(y_hat, label='Predicted')
plt.fill_between(range(len(time_series_data), len(time_series_data) + len(y_hat)),
                 yhat_conf_int.iloc[:, 0], yhat_conf_int.iloc[:, 1], color='gray', alpha=0.3, label='95% Confidence Interval')
plt.legend()
plt.show()
```

```python
model = ARIMA(time_series_data, order =(1, 1, 5))
model_fit = model.fit()
print(model_fit.summary())

forecast = model_fit.get_forecast(steps=27)
y_hat = forecast.predicted_mean
yhat_conf_int = forecast.conf_int(alpha=0.05)

plt.plot(test["Mean_Price_per_kg"])
plt.plot(y_hat, label='Predicted')
plt.fill_between(range(len(time_series_data), len(time_series_data) + len(y_hat)),
                 yhat_conf_int.iloc[:, 0], yhat_conf_int.iloc[:, 1], color='gray', alpha=0.3, label='95% Confidence Interval')
plt.legend()
plt.show()
```

## Evaluate the best model

```python
def compare_models(models, test):
    results = {}
    for i, model in enumerate(models, start=1):
        # Fit the model
        model_fit = model.fit()
        
        # Generate forecasts
        forecast = model_fit.forecast(steps=len(test))
        
        # Calculate the squared difference between forecast and test data
        diff_squared = (test["Mean_Price_per_kg"] - forecast) ** 2
        
        # Sum of squared differences
        sum_squared_diff = diff_squared.sum()
        print(sum_squared_diff)
        # Store the sum of squared differences for each model
        results[f"Model {i}"] = sum_squared_diff
  
    # Find the model with the smallest sum of squared differences
    best_model = min(results, key=results.get)

    return best_model

```

## Evaluate the best model with different time data

```python
train = new_df[new_df['Date'].dt.year < 2016]
new_test = new_df[new_df['Date'].dt.year >= 2016]
time_series_data_test = train['Mean_Price_per_kg']

model1 = ARIMA(time_series_data_test, order =(1, 1, 4))
model2 = SARIMAX(time_series_data_test, order=(1, 1, 5), seasonal_order=(1, 1, 1, 12))
model3 = ARIMA(time_series_data_test, order =(1, 1, 5))
model4 = SARIMAX(time_series_data_test, order=(1, 1, 2), seasonal_order=(1, 1, 1, 12))
model5 = SARIMAX(time_series_data_test, order=(2, 1, 2), seasonal_order=(1, 1, 1, 12))

models = [model1, model2, model3, model4, model5]


compare_models(models, new_test)
```

```python
time_series_data_new = new_df['Mean_Price_per_kg']

model= SARIMAX(time_series_data_new[:-1] , order=(1, 1, 5), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit()
print(model_fit.summary())


forecast = model_fit.forecast(steps=24)

confvalue = model_fit.get_forecast(steps=24)
yhat_conf_int = confvalue.conf_int(alpha=0.05)


plt.plot(time_series_data_new, label = "Past")
plt.plot(forecast, label='Predicted Future')
plt.fill_between(forecast.index, yhat_conf_int.iloc[:, 0], yhat_conf_int.iloc[:, 1], color='gray', alpha=0.3, label='95% Confidence Interval')
plt.legend()
plt.show()

plt.plot(forecast, label='Predicted')
plt.fill_between(forecast.index, yhat_conf_int.iloc[:, 0], yhat_conf_int.iloc[:, 1], color='gray', alpha=0.3, label='95% Confidence Interval')
plt.title("95% Confidence Interval")
plt.show()
```

## Final forecast

```python

model = SARIMAX(time_series_data_new[:-1], order=(1, 1, 5), seasonal_order=(1, 1, 1, 12))
model_fit = model.fit()

forecast_values = model_fit.forecast(steps=24)

last_date = new_df['Date'].max()
forecast_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=24, freq='ME')

confvalue = model_fit.get_forecast(steps=24)
conf = confvalue.conf_int(alpha = 0.05)

lower_bound = conf.iloc[:, 0]
upper_bound = conf.iloc[:, 1]


forecast_df = pd.DataFrame({'Date': forecast_dates, 'Mean_Price_per_kg': forecast_values, 'Lower Bound': lower_bound, 'Upper bound': upper_bound })

print(forecast_df)
```
